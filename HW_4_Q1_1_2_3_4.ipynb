```python
#importing libraries for numpy, plot and pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing breast cancer dataset
from sklearn.datasets import load_breast_cancer
breast = load_breast_cancer()
```

```python
#loading the input variables into a matrix
breast_data = breast.data
breast_data.shape
```

```python
#converting the input variables matrix into a 2D panda array
breast_input = pd.DataFrame(breast_data)
breast_input.head()
```

```python
#loading the outputs into a matrix
breast_labels = breast.target
breast_labels.shape
```

```python
breast_labels
```

```python
#reshaping the output row matrix into a column
labels = np.reshape(breast_labels,(569,1))
```

```python
#concatenating the input variables and outputs into the same 2D array to create the final dataset
final_breast_data = np.concatenate([breast_data,labels],axis=1)
```

```python
final_breast_data.shape
```

```python
#loading the feature names for the input variables
breast_dataset = pd.DataFrame(final_breast_data)
features = breast.feature_names
features
```

```python
features_labels = np.append(features,'label')
```

```python
breast_dataset.columns = features_labels
```

```python
breast_dataset
```

```python
breast_dataset['label'].replace(0, 'Benign',inplace=True)
breast_dataset['label'].replace(1, 'Malignant',inplace=True)
```

```python
breast_dataset
```

```python
#importing standard scalar and standarding the input values
from sklearn.preprocessing import StandardScaler
# Separating out the features
x = breast_dataset.loc[:, features].values
# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```python
#importing Principal Component Analysis
from sklearn.decomposition import PCA
pca = PCA(n_components=9)
principalComponents = pca.fit_transform(x)
principalComponents = pd.DataFrame(principalComponents)
principalComponents
```

```python
#Setting the principal components as the input variables
X = principalComponents.iloc[:, [0, 1, 2, 3, 4, 5, 6, 7, 8]].values
Y = breast_dataset.iloc[:, 30].values
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8, test_size=0.2, random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
grid = GridSearchCV(SVC(), param_grid, refit=True, cv=5)
grid.fit(X_train, Y_train)
```

```python
grid.best_params_
```

```python
Y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print("Confusion Matrix:\n", confusion_matrix(Y_test, Y_pred))
print("\n")
print("Classification Report:\n",classification_report(Y_test, Y_pred))
```

```python
from sklearn import metrics
pos_label=['Benign', 'Malignant']
print("Accuracy:",metrics.accuracy_score(Y_test, Y_pred))
print("Precision:",metrics.precision_score(Y_test, Y_pred, pos_label= "Malignant"))
print("Recall:",metrics.recall_score(Y_test, Y_pred, pos_label= "Malignant"))
```

```python
# Run SVM with rbf kernel and C=100.0
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
grid = GridSearchCV(SVC(), param_grid, refit=True, cv=5)
grid.fit(X_train, Y_train)
```

```python
# instantiate classifier with rbf kernel and C=10
svc=SVC(C=10.0)
svc.fit(X_train,Y_train)
# make predictions on test set
Y_pred=svc.predict(X_test)
# compute and print accuracy score
print('Model accuracy score with rbf kernel and C=10.0 : {}'. format(accuracy_score(Y_test, Y_pred)))
```

```python
# SVM with linear kernel
# SVM with linear kernel and C=100.0
linear_svc=SVC(kernel='linear', C=100.0)
linear_svc.fit(X_train,Y_train)
# make predictions on test set
Y_pred_test=linear_svc.predict(X_test)
# compute and print accuracy score
print('Model accuracy score with linear kernel and C=100.0 : {0:0.4f}'. format(accuracy_score(Y_test, Y_pred_test)))
```

```python
#SVM with polynomial kernel and C=1000.0
poly_svc100=SVC(kernel='poly', C=1000.0)
# fit classifier to training set
poly_svc100.fit(X_train, Y_train)
# make predictions on test set
Y_pred=poly_svc100.predict(X_test)
# compute and print accuracy score
print('Model accuracy score with polynomial kernel and C=1000.0 : {0:0.4f}'. format(accuracy_score(Y_test, Y_pred)))
```

```python
# Run SVM with sigmoid kernel
# instantiate classifier with sigmoid kernel and C=100.0
sigmoid_svc100=SVC(kernel='sigmoid', C=100.0)
# fit classifier to training set
sigmoid_svc100.fit(X_train,Y_train)
# make predictions on test set
Y_pred=sigmoid_svc100.predict(X_test)
# compute and print accuracy score
print('Model accuracy score with sigmoid kernel and C=100.0 : {0:0.4f}'. format(accuracy_score(Y_test, Y_pred)))
```
